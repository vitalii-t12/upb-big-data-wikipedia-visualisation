```python
from mwviews.api import PageviewsClient

# USER_AGENT = 'UPB-project (veryprettystudent@gmail.com)'
USER_AGENT = 'MyWikiApp/1.0 (myemail@example.com)'

# Initialize the Pageviews Client
client = PageviewsClient(USER_AGENT)

# Specify the project (e.g., English Wikipedia)
project = 'en.wikipedia'

# Fetch the top articles for a given project
# Optionally, specify the date, if not it will return the latest data
try:
    top_articles = client.top_articles(project)
    print(top_articles)
    for article in top_articles[:10]:  # Display the top 10 articles
      print(f"Title: {article['article']}, Views: {article['views']}, Rank: {article['rank']}")
except Exception as e:
    print(f"Error fetching top articles: {e}")
```

# Top pages per country

```python
import requests

def get_top_pages_per_country(project, access, year, month, day, country):
    url = f"https://wikimedia.org/api/rest_v1/metrics/pageviews/top-per-country/{project}/{access}/{year}/{month}/{day}"
    headers = {
        "User-Agent": "YourAppName/1.0 (contact@example.com)"
    }

    try:
        # Make the request
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Raise an error for bad HTTP status codes

        # Parse the JSON response
        data = response.json()
        country_data = [
            entry for entry in data.get("items", [])
            if entry.get("country") == country
        ]

        # Display results
        if country_data:
            print(f"Top pages in {country} on {year}-{month}-{day}:")
            for article in country_data[0].get("articles", [])[:10]:  # Limit to top 10
                print(f"Title: {article['article']}, Views: {article['views']}")
        else:
            print(f"No data found for {country} on {year}-{month}-{day}.")

    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")

# Example usage
get_top_pages_per_country(
    project="en.wikipedia",
    access="all-access",
    year="2023",
    month="12",
    day="01",
    country="US"  # Country code (e.g., US for the United States)
)
```

```python
import requests

def get_top_articles(project, access, year, month, day):
    url = f"https://wikimedia.org/api/rest_v1/metrics/pageviews/top/{project}/{access}/{year}/{month}/{day}"
    headers = {
        "User-Agent": "YourAppName/1.0 (contact@example.com)"
    }

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()

        data = response.json()
        articles = data['items'][0]['articles']
        print(f"Top articles on {year}-{month}-{day}:")
        for article in articles[:10]:  # Top 10 articles
            print(f"Title: {article['article']}, Views: {article['views']}")

    except requests.exceptions.RequestException as e:
        print(f"An error occurred: {e}")

# Example usage
get_top_articles(
    project="en.wikipedia",
    access="all-access",
    year="2023",
    month="12",
    day="01"
)
```
